```python
import pandas as pd
df = pd.read_csv('apnastudent.csv')
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('price', axis ='columns'), df.price)
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
```

```python
df.isnull()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
df1 = pd.DataFrame({'original_value':y_test, 'predict_value':y_pred})
df1
```

```python
df1.to_csv('RMSE_predicted.csv')
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae_error = mae(y_test,y_pred)
```
